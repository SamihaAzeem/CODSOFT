```python
from tkinter import *
from tkinter import ttk
import csv
import pickle
from tkinter import filedialog

root = Tk()
root.title('Contact Book')
root.geometry("600x700")

# Add some style
style = ttk.Style()
#Pick a theme
style.theme_use("default")


# Create Treeview Frame
tree_frame = Frame(root)
tree_frame.pack(pady=20)

# Treeview Scrollbar
tree_scroll = Scrollbar(tree_frame)
tree_scroll.pack(side=RIGHT, fill=Y)

# Create Treeview
my_tree = ttk.Treeview(tree_frame, yscrollcommand=tree_scroll.set, selectmode="extended")
# Pack to the screen
my_tree.pack()

#Configure the scrollbar
tree_scroll.config(command=my_tree.yview)

# Define Our Columns
my_tree['columns'] = ("Name", "Phone", "Email", "Address")

# Formate Our Columns
my_tree.column("#0", width=0, stretch=NO)
my_tree.column("Name", anchor=W, width=140)
my_tree.column("Phone", anchor=CENTER, width=100)
my_tree.column("Email", anchor=W, width=140)
my_tree.column("Address", anchor=W, width=100)

# Create Headings 
my_tree.heading("#0", text="", anchor=CENTER)
my_tree.heading("Name", text="Name", anchor=CENTER)
my_tree.heading("Phone", text="Phone", anchor=CENTER)
my_tree.heading("Email", text="Email", anchor=CENTER)
my_tree.heading("Address", text="Address", anchor=CENTER)

# Add Dummy Data
# data = [
#     ['Siddharth Nigam', '369854712', 'sid@gmail.com', 'H#12 St13'],
#     ['Gaurav Patil', '521155222', 'g2@gmail.com', 'H#2 St13'],
#     ['Abhishek Nikam', '78945614', 'a3@gmail.com', 'H#3 St13'],
#     ['Sakshi Gaikwad', '58745246', 's4@gmail.com', 'H#4 St13'],
#     ['Mohit Paul', '5846975', 'sid5@gmail.com', 'H#5 St13'],
#     ['Karan Patel', '5647892', 'si6@gmail.com', 'H#6 St13'],
#     ['Ganesh Pawar', '85967412', 'sid99@gmail.com', 'H#9 St13']
# ]

# global count
# count=0

# for record in data:
#     my_tree.insert(parent='', index='end', iid=count, text="", values=(record[0], record[1], record[2], record[3])) 
#     count += 1



add_frame = Frame(root)
add_frame.pack(pady=20)

#Labels
nl = Label(add_frame, text="Name")
nl.grid(row=0, column=0)

il = Label(add_frame, text="Phone")
il.grid(row=0, column=1)

tl = Label(add_frame, text="Email")
tl.grid(row=0, column=2)

tl = Label(add_frame, text="Address")
tl.grid(row=0, column=3)



#Entry boxes
name_box = Entry(add_frame)
name_box.grid(row=1, column=0)

phone_box = Entry(add_frame)
phone_box.grid(row=1, column=1)

email_box = Entry(add_frame)
email_box.grid(row=1, column=2)

address_box = Entry(add_frame)
address_box.grid(row=1, column=3)

# Add Record
def add_record():
    global count
    my_tree.insert(parent='', index='end', iid=count, text="", values=(name_box.get(), phone_box.get(), email_box.get(),address_box.get()))
    count += 1

    # Clear the boxes
    name_box.delete(0, END)
    phone_box.delete(0, END)
    email_box.delete(0, END)
    address_box.delete(0, END)

# Remove all records
def remove_all():
    for record in my_tree.get_children():
        my_tree.delete(record)

# Remove one selected
def remove_one():
    x = my_tree.selection()[0]
    my_tree.delete(x)

# Remove many selected
def remove_many():
    x = my_tree.selection()
    for record in x:
        my_tree.delete(record)

# Select Record
def select_record():
    # Clear entry boxes
    name_box.delete(0, END)
    phone_box.delete(0, END)
    email_box.delete(0, END)
    address_box.delete(0, END)

    # Grab record number
    selected = my_tree.focus()
    # Grab record values
    values = my_tree.item(selected, 'values')

    #temp_label.config(text=values[0])

    # output to entry boxes
    name_box.insert(0, values[0])
    phone_box.insert(0, values[1])
    email_box.insert(0, values[2])
    address_box.insert(0, values[3])


# Save updated record
def update_record():
    # Grab record number
    selected = my_tree.focus()
    # Save new data
    my_tree.item(selected, text="", values=(name_box.get(), phone_box.get(), email_box.get(), address_box.get()))

    # Clear entry boxes
    name_box.delete(0, END)
    phone_box.delete(0, END)
    email_box.delete(0, END)
    address_box.delete(0, END)
    

# Function to exit the application
def exit_app():
    root.destroy()
    
    
def search_records():
    search_term = search_entry.get()
    for item in my_tree.get_children():
        values = my_tree.item(item, 'values')
        if search_term.lower() in [str(value).lower() for value in values]:
            my_tree.selection_set(item)
        else:
            my_tree.selection_remove(item)

def save_data():
    data_to_save = []
    for item in my_tree.get_children():
        data = my_tree.item(item, 'values')
        data_to_save.append(data)
    
    # Prompt the user for the file name and location
    file_path = filedialog.asksaveasfilename(defaultextension=".pkl", filetypes=[("Pickle Files", "*.pkl")])
    
    if file_path:
        with open(file_path, 'wb') as file:
            pickle.dump(data_to_save, file)

# Function to load data from a file using pickle and populate the Treeview
def load_data():
    file_path = filedialog.askopenfilename(filetypes=[("Pickle Files", "*.pkl")])
    if file_path:
        try:
            with open(file_path, 'rb') as file:
                saved_data = pickle.load(file)
                for record in saved_data:
                    my_tree.insert('', 'end', values=record)
        except FileNotFoundError:
            pass


search_label = Label(root, text="Search:")
search_label.pack(pady=10)

search_entry = Entry(root)
search_entry.pack(pady=5)

## Create the first row of buttons
button_frame1 = Frame(root)
button_frame1.pack(pady=10)

select_button = Button(button_frame1, text="Select Record", command=select_record)
update_button = Button(button_frame1, text="Save Record", command=update_record)
add_record = Button(button_frame1, text="Add Record", command=add_record)
remove_all = Button(button_frame1, text="Remove All Records", command=remove_all)
remove_one = Button(button_frame1, text="Remove One Selected", command=remove_one)

select_button.grid(row=0, column=0, padx=5)
update_button.grid(row=0, column=1, padx=5)
add_record.grid(row=0, column=2, padx=5)
remove_all.grid(row=0, column=3, padx=5)
remove_one.grid(row=0, column=4, padx=5)

# Create the second row of buttons
button_frame2 = Frame(root)
button_frame2.pack(pady=10)

remove_many = Button(button_frame2, text="Remove Many Selected", command=remove_many)
save_button = Button(button_frame2, text="Save Data", command=save_data)
load_button = Button(button_frame2, text="Load Data", command=load_data)

remove_many.grid(row=0, column=0, padx=5)
save_button.grid(row=0, column=1, padx=5)
load_button.grid(row=0, column=2, padx=5)



exit_button = Button(root, text="Exit", command=exit_app)
exit_button.pack(pady=5)
search_entry.bind('<Return>', lambda event=None: search_records())



root.mainloop()

```
