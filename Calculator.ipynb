```python
import tkinter as tk

field_text = " "

def add_field(text):
    global field_text
    field_text = field_text + str(text)
    field.delete(1.0, tk.END)
    field.insert(tk.END, field_text)
    
def calculate():
    result = str(eval(field_text))
    field.delete(1.0, tk.END)
    field.insert(tk.END, result)
    
def clear():
    global field_text 
    field_text = " "
    field.delete(1.0, tk.END)
    
def backspace():
    global field_text
    field_text = field_text[:-1]
    field.delete(1.0, tk.END)
    field.insert(tk.END, field_text)

root = tk.Tk()  # Create the main window
root.geometry("310x300")

#title of our window
root.title("Calculator")

field = tk.Text(root, height=2, width=21,borderwidth=5, font=("Times New Roman", 20))
field.grid(row=1, column=1, columnspan=4)

btn_1 = tk.Button(root, text="1", command=lambda: add_field(1), width=5, font=("Times New Roman", 14))
btn_1.grid(row=4, column=1)

btn_2 = tk.Button(root, text="2", command=lambda: add_field(2), width=5, font=("Times New Roman", 14))
btn_2.grid(row=4, column=2)

btn_3 = tk.Button(root, text="3", command=lambda: add_field(3), width=5, font=("Times New Roman", 14))
btn_3.grid(row=4, column=3)

btn_4 = tk.Button(root, text="4", command=lambda: add_field(4), width=5, font=("Times New Roman", 14))
btn_4.grid(row=3, column=1)

btn_5 = tk.Button(root, text="5", command=lambda: add_field(5), width=5, font=("Times New Roman", 14))
btn_5.grid(row=3, column=2)

btn_6 = tk.Button(root, text="6", command=lambda: add_field(6), width=5, font=("Times New Roman", 14))
btn_6.grid(row=3, column=3)

btn_7 = tk.Button(root, text="7", command=lambda: add_field(7), width=5, font=("Times New Roman", 14))
btn_7.grid(row=2, column=1)

btn_8 = tk.Button(root, text="8", command=lambda: add_field(8), width=5, font=("Times New Roman", 14))
btn_8.grid(row=2, column=2)

btn_9 = tk.Button(root, text="9", command=lambda: add_field(9), width=5, font=("Times New Roman", 14))
btn_9.grid(row=2, column=3) # 7 1

btn_0 = tk.Button(root, text="0", command=lambda: add_field(0), width=5, font=("Times New Roman", 14))
btn_0.grid(row=5, column=1)

btn_plus = tk.Button(root, text="+", command=lambda: add_field("+"), width=5, font=("Times New Roman", 14))
btn_plus.grid(row=4, column=4)

btn_minus = tk.Button(root, text="-", command=lambda: add_field("-"), width=5, font=("Times New Roman", 14))
btn_minus.grid(row=5, column=4)

btn_mul = tk.Button(root, text="*", command=lambda: add_field("*"), width=5, font=("Times New Roman", 14))
btn_mul.grid(row=3, column=4)

btn_div = tk.Button(root, text="/", command=lambda: add_field("/"), width=5, font=("Times New Roman", 14))
btn_div.grid(row=6, column=4) # 2 4

btn_clear = tk.Button(root, text="C", command=lambda: clear(), width=5, font=("Times New Roman", 14))
btn_clear.grid(row=5, column=3)

btn_deci = tk.Button(root, text=".", command=lambda: add_field("."), width=5, font=("Times New Roman", 14))
btn_deci.grid(row=5, column=2)

btn_opn_paran = tk.Button(root, text="(", command=lambda: add_field("("), width=5, font=("Times New Roman", 14))
btn_opn_paran.grid(row=6, column=1)

btn_close_paran = tk.Button(root, text= ")", command=lambda: add_field( ")"), width=5, font=("Times New Roman", 14))
btn_close_paran.grid(row=6, column=2)

btn_equal = tk.Button(root, text="=", command=lambda: calculate(), width=5, font=("Times New Roman", 14))
btn_equal.grid(row=6, column=3)

btn_backspace = tk.Button(root, text="âŒ«", command=lambda: backspace(), width=5, font=("Times New Roman", 14))
btn_backspace.grid(row=2, column=4)

root.config(bg="white")
root.mainloop()  
```
