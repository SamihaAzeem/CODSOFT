```python
from tkinter import *
from tkinter.font import Font
from tkinter import filedialog
import pickle

root = Tk()
root.title('To Do List')
root.geometry("450x500")


# Define our Font
my_font = Font(
    family="Comic Sans MS",
    size=20,
    weight="bold")

# Creat frame
my_frame = Frame(root)
my_frame.pack(pady=10)

# Create listbox
my_list = Listbox(my_frame,
    font=my_font,
    width=25,
    height=5,
    bg="SystemButtonFace",
    bd=0,
    fg="#5C5470",
    highlightthickness=0,
    selectbackground="#9F91CC",
    activestyle="none")

my_list.pack(side=LEFT, fill=BOTH)

# Create scrollbar
my_scrollbar = Scrollbar(my_frame)
my_scrollbar.pack(side=RIGHT, fill=BOTH)

# Add scrollbar
my_list.config(yscrollcommand=my_scrollbar.set)
my_scrollbar.config(command=my_list.yview)

# create entry box to add items to the list
my_entry = Entry(root, font=("Comic Sans MS", 20), width=22,fg="#5C5470")
my_entry.pack(pady=20)

# Create a button frame
button_frame = Frame(root)
button_frame.pack(pady=20)

# FUNCTIONS
def delete_item():
	my_list.delete(ANCHOR)

def add_item():
    my_list.insert(END, my_entry.get())
    my_entry.delete(0, END)

def cross_off_item():
    # Cross off item
    my_list.itemconfig(
    my_list.curselection(),
    fg="#B9B4C7")
    # Get rid of selection bar
    my_list.selection_clear(0, END)

def uncross_item():
    # Cross off item
    my_list.itemconfig(
        my_list.curselection(),
        fg="#5C5470")
    # Get rid of selection bar
    my_list.selection_clear(0, END)
    
def update_item():
    # Get the currently selected item's index
    selected_index = my_list.curselection()

    # Check if an item is selected
    if selected_index:
        # Get the text from the entry widget
        new_text = my_entry.get()

        # Update the selected item with the new text
        my_list.delete(selected_index[0])  # Delete the selected item
        my_list.insert(selected_index[0], new_text)  # Insert the new text at the same index
        my_entry.delete(0, END)


def delete_crossed():
    count = 0
    while count < my_list.size():
        if my_list.itemcget(count, "fg") == "#B9B4C7":
            my_list.delete(my_list.index(count))

        else: 
            count += 1

def save_list():
    file_name = filedialog.asksaveasfilename(
        initialdir="C:/Users/ABD/Desktop/Untitled Folder/PYTHON TASKS",
        title="Save File",
        filetypes=(
            ("Dat Files", "*.dat"), 
            ("All Files", "*.*"))
        )
    if file_name:
        if file_name.endswith(".dat"):
            pass
        else:
            file_name = f'{file_name}.dat'

        # Delete crossed off items before saving
        count = 0
        while count < my_list.size():
            if my_list.itemcget(count, "fg") == "#B9B4C7":
                my_list.delete(my_list.index(count))

            else: 
                count += 1

        # grab all the stuff from the list
        stuff = my_list.get(0, END)

        # Open the file
        output_file = open(file_name, 'wb')

        # Actually add the stuff to the file
        pickle.dump(stuff, output_file)


def open_list():
    file_name = filedialog.askopenfilename(
        initialdir="C:/Users/ABD/Desktop/Untitled Folder/PYTHON TASKS",
        title="Open File",
        filetypes=(
            ("Dat Files", "*.dat"), 
            ("All Files", "*.*"))
        )

    if file_name:
        # Delete currently open list
        my_list.delete(0, END)

        # Open the file
        input_file = open(file_name, 'rb')

        # Load the data from the file
        stuff = pickle.load(input_file)

        # Output stuff to the screen
        for item in stuff:
            my_list.insert(END, item)


def on_entry_click(event):
    if my_entry.get() == "Enter your ToDos...":
        my_entry.delete(0, "end")
        my_entry.config(fg="#5C5470")  # Change text color to black

def on_entry_leave(event):
    if not my_entry.get():
        my_entry.insert(0, "Enter your ToDos...")
        my_entry.config(fg="gray")  # Change text color to gray


def clear_list():
    my_list.delete(0, END)
    
    
my_entry.insert(0, "Enter your ToDos...")
my_entry.bind("<FocusIn>", on_entry_click)
my_entry.bind("<FocusOut>", on_entry_leave)
my_entry.pack()


# Add some buttons
delete_button = Button(button_frame, text="Delete Item",bg="#B9B4C7", command=delete_item)
add_button = Button(button_frame, text="Add Item",bg="#B9B4C7", command=add_item)
cross_off_button = Button(button_frame, text="Cross Off Item",bg="#B9B4C7", command=cross_off_item)
uncross_button = Button(button_frame, text="Uncross Item",bg="#B9B4C7", command=uncross_item)
delete_crossed_button = Button(button_frame, text="Delete Crossed",bg="#B9B4C7", command=delete_crossed)
update_button = Button(button_frame, text="Update Item",bg="#B9B4C7",command=update_item)
save_button = Button(button_frame, text="Save List",bg="#3F2E3E",fg="white",command=save_list)
open_button = Button(button_frame, text="Open List",bg="#A78295",fg="white",command=open_list)
clear_button = Button(button_frame, text="Clear List",bg="#331D2C",fg="white",command=clear_list)

delete_button.grid(row=0, column=1)
add_button.grid(row=0, column=0, padx=20)
update_button.grid(row=0,column=2, padx=20)

cross_off_button.grid(row=1, column=0, pady=10)
uncross_button.grid(row=1, column=1, padx=20)
delete_crossed_button.grid(row=1, column=2)

save_button.grid(row=2, column=0) 
open_button.grid(row=2, column=1)  
clear_button.grid(row=2, column=2)  

root.mainloop()
```
