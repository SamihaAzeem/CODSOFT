```python
import random ,string
import pyperclip as pc
from tkinter import *



#Initialize Window
 
root =Tk()
root.geometry("500x520") 
 
#title of our window
root.title("Random Password Generator")

msg = Label(root, text ='Welcome to Password Generator', font = "Comfortaa 18 bold",fg="deepskyblue4").pack(pady=10)
 
output_pass = StringVar()
 

# -------------------  Random Password generator function
def randPassGen():
    password=''
    words=[random.choice(string.ascii_letters) for i in range(pass_words_len.get())]
    numbers=[random.choice(string.digits) for i in range(pass_num_len.get())]
    symbol=[random.choice(string.punctuation) for i in range(pass_symbol_len.get())]

    key = words+numbers+symbol

    random.shuffle(key)

    for i in key:
        password = password+i

    output_pass.set(password)
    
    
def copyPass():
    pc.copy(output_pass.get())
    
    
#Initialize a Label to display the User Input

pass_words = Label(root, text = 'Enter how many letters: ', font = 'Comfortaa 12 bold', fg="burlywood4").pack(pady=10)  
pass_words_len = IntVar() #integer variable to store the input of words length of the password wanted
length_words = Spinbox(root, from_ = 2, to_ = 32 , textvariable = pass_words_len , width = 24, font='Comfortaa 16',fg="deepskyblue4").pack()


pass_num = Label(root, text = 'Enter how many numbers: ', font = 'Comfortaa 12 bold', fg="burlywood4").pack(pady=10) 
pass_num_len = IntVar() #integer variable to store the input of number length of the password wanted
length_num = Spinbox(root, from_ = 2, to_ = 32 , textvariable = pass_num_len , width = 24, font='Comfortaa 16',fg="deepskyblue4").pack()



pass_symbol = Label(root, text = 'Enter how many symbols: ', font = 'Comfortaa 12 bold', fg="burlywood4").pack(pady=10)  
pass_symbol_len = IntVar() #integer variable to store the input of symbols length of the password wanted
length_symbol = Spinbox(root, from_ = 0, to_ = 32 , textvariable = pass_symbol_len , width = 24, font='Comfortaa 16',fg="deepskyblue4").pack()


   
#Generate password button
 
Button(root, command = randPassGen, text = "Generate Password", font="Comfortaa 12 bold", bg='deepskyblue4', fg='white', activebackground="burlywood3", padx=5, pady=5 ).pack(pady= 30)
 
pass_label = Label(root, text = 'Random Generated Password', font = 'Comfortaa 12 bold',fg='deepskyblue4').pack(pady="10")
Entry(root , textvariable = output_pass, width = 30, font='Comfortaa 12 bold',fg='burlywood4').pack(pady="10")

#Copy to clipboard button
 
Button(root, text = 'Copy to Clipboard', command = copyPass, font="Comfortaa 12 bold", bg='deepskyblue4', fg='white', activebackground="burlywood3", padx=5, pady=5 ).pack(pady= 10)
 

root.mainloop()  #to bundle pack the code for tkinter
```
