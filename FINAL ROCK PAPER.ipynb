```python
import tkinter as tk
import random
from PIL import ImageTk, Image

user_score = 0
computer_score = 0


# Function to determine the winner
def play(user_choice):
    global user_score, computer_score,img

    # Generate a random choice for the computer
    choices = ["rock", "paper", "scissors"]
    computer_choice = random.choice(choices)

    # Update the labels to display choices
    user_choice_label.config(text=f"Your choice: {user_choice}")
    computer_choice_label.config(text=f"Computer's choice: {computer_choice}")

    # Determine the winner
    if user_choice == computer_choice:
        result_label.config(text="It's a Tie!")
        frame = tk.Frame(root, width=400, height=400)
        frame.pack()
        frame.place(anchor='center', relx=0.5, rely=0.7)

        # Create an object of tkinter ImageTk
        img = ImageTk.PhotoImage(Image.open("tie.png"))

        # Create a Label Widget to display the text or Image
        label = tk.Label(frame, image = img)
        label.pack()
    elif (
        (user_choice == "rock" and computer_choice == "scissors")
        or (user_choice == "paper" and computer_choice == "rock")
        or (user_choice == "scissors" and computer_choice == "paper")
    ):
        result_label.config(text="You Win!")
        
        frame = tk.Frame(root, width=400, height=400)
        frame.pack()
        frame.place(anchor='center', relx=0.5, rely=0.7)

        # Create an object of tkinter ImageTk
        img = ImageTk.PhotoImage(Image.open("yay.png"))

        # Create a Label Widget to display the text or Image
        label = tk.Label(frame, image = img)
        label.pack()
        user_score += 1
    else:
        result_label.config(text="Computer Wins!")
        frame = tk.Frame(root, width=400, height=400)
        frame.pack()
        frame.place(anchor='center', relx=0.5, rely=0.7)

        # Create an object of tkinter ImageTk
        img = ImageTk.PhotoImage(Image.open("sad.png"))

        # Create a Label Widget to display the text or Image
        label = tk.Label(frame, image = img)
        label.pack()
        computer_score += 1

    # Update the scores
    user_score_label.config(text=f"Your Score: {user_score}")
    computer_score_label.config(text=f"Computer's Score: {computer_score}")

# Function to reset the scores
def reset_scores():
    global user_score, computer_score, img
    user_score = 0
    computer_score = 0
    user_score_label.config(text="Your Score: 0")
    computer_score_label.config(text="Computer's Score: 0")
    img=None

# Create the main window
root = tk.Tk()
root.title('Rock, Paper, Scissors')
root.geometry("600x700")  # Increased height to fit Reset Scores button
root.config(bg="white")

# Create labels and buttons
user_choice_label = tk.Label(root, text="", font=("Helvetica", 14))
user_choice_label.pack(pady=10)

computer_choice_label = tk.Label(root, text="", font=("Helvetica", 14))
computer_choice_label.pack(pady=10)

result_label = tk.Label(root, text="", font=("Helvetica", 18))
result_label.pack(pady=20)

user_score_label = tk.Label(root, text="Your Score: 0", font=("Helvetica", 14))
user_score_label.pack()

computer_score_label = tk.Label(root, text="Computer's Score: 0", font=("Helvetica", 14))
computer_score_label.pack()

# Create a frame for buttons
button_frame = tk.Frame(root, bg="white")
button_frame.pack()

button_frame2 = tk.Frame(root, bg="white")
button_frame2.pack()

rock_button = tk.Button(button_frame, text="Rock", command=lambda: play("rock"))
paper_button = tk.Button(button_frame, text="Paper", command=lambda: play("paper"))
scissors_button = tk.Button(button_frame, text="Scissors", command=lambda: play("scissors"))
reset_button = tk.Button(button_frame2, text="Reset Scores", command=reset_scores)
exit_button = tk.Button(button_frame2, text="Exit",bg="red",fg="white",font='Helvetica 12 bold',command=root.destroy)


rock_button.pack(side=tk.LEFT, padx=15,pady=10)
paper_button.pack(side=tk.LEFT, padx=15,pady=10)
scissors_button.pack(side=tk.LEFT, padx=15,pady=10)
reset_button.pack(side=tk.LEFT, padx=15,pady=10)
exit_button.pack(side=tk.LEFT, padx=15,pady=10)

root.mainloop()
```
